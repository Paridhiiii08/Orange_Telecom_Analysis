```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath='Orange_Telecom_Churn_Data.csv'
```

```python
data =pd.read_csv(filepath)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data= data.drop(columns=['phone_number'])
```

```python
data['intl_plan']=data['intl_plan'].map({'yes':1, 'no':0})

data['voice_mail_plan'] = data['voice_mail_plan'].map({'yes': 1, 'no': 0})
```

```python
data= data.drop_duplicates()
```

```python
data.isnull().sum()
```

```python
data['intl_plan'].unique()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='churned', data=data)
plt.title('Churn Rate')
plt.xticks([0, 1], ['Not Churned', 'Churned'])
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Churn by International Plan
sns.barplot(x='intl_plan', y='churned', data=data, estimator=lambda x: x.mean())
plt.title('Churn Rate by International Plan')
plt.xticks([0, 1], ['No Intl Plan', 'Intl Plan'])
plt.show()

```

```python
sns.barplot(x='voice_mail_plan', y='churned', data=data, estimator=lambda x: x.mean())
plt.title('Churn Rate by Voicemail Plan')
plt.xticks([0, 1], ['No Voicemail Plan', 'Voicemail Plan'])
plt.show()
```

```python
customer_service_churn = data.groupby('number_customer_service_calls')['churned'].mean().reset_index()
sns.lineplot(x='number_customer_service_calls', y='churned', data=customer_service_churn, marker='o')
plt.title('Churn Rate by Customer Service Calls')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Churn Rate')
plt.show()
```

```python
usage_columns = ['total_day_minutes', 'total_eve_minutes', 'total_night_minutes', 'total_intl_minutes']
plt.figure(figsize=(12, 8))
for i, column in enumerate(usage_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```
